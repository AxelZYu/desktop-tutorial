```python
import pandas as pd
'''Загрузим данные по населению г. Москва'''
people = pd.read_csv("https://video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv", delimiter=";")
people = people.set_index(["Year", "Period"])#Укажим индекс на  две колонки
print (people.head())# Посмотрим из чего сотоит набор этих данных
MCHS = pd.read_csv("https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv", delimiter=";")
'''Загрузим данные по вызовам пожарной службы'''
MCHS = MCHS.set_index(["AdmArea", "Year", "Month"])#Укажим индекс на  три колонки
print (MCHS.head())
MCHS = MCHS.loc["Центральный административный округ"]#Используем только те данные которые относятся к "Центральный административный округ"
MCHS.index.names = ["Year", "Period"]#Перенаименуем столбец для дальнейшего объединения двух фреймов данных
print (MCHS.head())
data = pd.merge(people, MCHS, left_index=True, right_index=True)#Объеденим данные
data = data.reset_index()#сбросим индекс
data = data.set_index("Calls")#Переназначим индекс на звонки
'''Отсортируем полученные данные и выведем минимальное(первое) значение'''
data = data.sort_index()
print (data)
'''Теперь получим данные по минимальному количеству вызовов и соответствюющее ему количество безработных'''
print (data["UnemployedMen"][0:1])
```

```python
import pandas as pd
people = pd.read_csv("https://video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv", delimiter=";")
people = people.set_index(["Year", "Period"])
MCHS = pd.read_csv("https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv", delimiter=";")
MCHS = MCHS.set_index(["AdmArea", "Year", "Month"])
MCHS = MCHS.loc["Центральный административный округ"]
MCHS.index.names = ["Year", "Period"]
data = pd.merge(people, MCHS, left_index=True, right_index=True)
data = data.reset_index()
data = data.set_index("Calls")
data = data.sort_index()
print (data["UnemployedMen"][0:1])
```
