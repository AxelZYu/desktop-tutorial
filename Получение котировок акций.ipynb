```python
import requests
import pandas as pd
from bs4 import BeautifulSoup#Загружаем библиотеки и получаем данные с сайта
r=requests.get ("http://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019")
html=BeautifulSoup(r.content)
data=[]#Создаем список
table = html.find('table', {"id":"marketDataList"})#Отбираем тэги с нужными данными
for tr in table.find_all("tr"):#перебераем все тэги с tr
    data.append([td.get_text() for td in tr.find_all("td")])#перебераем все заданные ячейки в найденных строках, берем из них текст и добовляем в созданный список
data=pd.DataFrame(data, columns=["Тикер", "Дата", "Сделки", "C/рост", "С/%", "Закрытие", "Открытие", "min", "max", "avg", "шт", "руб", "Всего"])#Передаем данные в дата фрейм и обозначаем колонки
data=data [data["Сделки"] !="N/A"]#Неучитываем значения не имеющих данных
data["С/%"]=data["С/%"].str.replace("−", "-").str.replace("%", "").astype(float)#В необходимом столбце прописываем читаемый -, убираем процент, меняем тип данных
data=data.set_index(["С/%"])#Указываем индекс
data=data.sort_index()#сортируем данный
print (data)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup#Загружаем библиотеки и получаем данные с сайта
r=requests.get ("http://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019")
html=BeautifulSoup(r.content)
data=[]#Создаем список
table = html.find('table', {"id":"marketDataList"})#Отбираем тэги с нужными данными
for tr in table.find_all("tr"):#перебераем все тэги с tr
    data.append([td.get_text() for td in tr.find_all("td")])#перебераем все заданные ячейки в найденных строках, берем из них текст и добовляем в созданный список
data=pd.DataFrame(data, columns=["Тикер", "Дата", "Сделки", "C/рост", "С/%", "Закрытие", "Открытие", "min", "max", "avg", "шт", "руб", "Всего"])#Передаем данные в дата фрейм и обозначаем колонки
data=data [data["Сделки"] !="N/A"]#Неучитываем значения не имеющих данных
data["С/%"]=data["С/%"].str.replace("−", "-").str.replace("%", "").astype(float)#В необходимом столбце прописываем читаемый -, убираем процент, меняем тип данных
data=data.set_index(["С/%"])#Указываем индекс
data=data.sort_index(ascending=False)#Обратим внимание на разную сортировку
print (data["Тикер"].head(1))
```
